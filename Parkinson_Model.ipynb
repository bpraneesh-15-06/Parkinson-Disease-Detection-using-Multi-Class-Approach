# FNN


```python
# MLP Classifier (Feedforward Neural Network)
mlp = MLPClassifier(max_iter=1000)
params_mlp = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.001, 0.01]
}
mlp_cv = RandomizedSearchCV(mlp, params_mlp, scoring='f1', n_iter=20, cv=5, random_state=42)
mlp_cv.fit(X_train_sm, y_train_sm)
mlp_best = mlp_cv.best_estimator_
y_pred_mlp = mlp_best.predict(X_test)
```

# Metrics Function


```python
# Evaluation Function
def print_metrics(name, y_true, y_pred):
    print(f"\n=== {name} ===")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
    print("Classification Report:\n", classification_report(y_true, y_pred))

print_metrics("KNN", y_test, y_pred_knn)
print_metrics("Feedforward Neural Network", y_test, y_pred_mlp)
```
